```python
# Create a new empty data frame
calibrated_data_formatted = pd.DataFrame(columns=['time', 'Process temperature', 'Ratio temperature'])
# For each line/row in the data frame that we created earlier by reading from the file
for line_number in range(3047):
    # Get the string of the current line/row
    current_line_string = calibrated_data.iloc[line_number,:].str
    # Get the data from above, turns out the following is the way
    current_line_string_list = current_line_string.split()[0]
    # replace , with . for decimal point in all the columns in the current line/row
    current_line_string_list = [current_entry.replace(',','.') for current_entry in current_line_string_list]
    # Add the current line/row data string to the new line in the new dataframe- CamTdata_10_with_emissivity_equals_1
    calibrated_data_formatted.loc[np.shape(calibrated_data_formatted.index)[0]] = current_line_string_list[:3]

    # Print some first and last rows to see what the data looks like
    if line_number < 3 or line_number > 7769:
        print(f'Data in line/row {line_number} is {current_line_string_list[:3]}')
```

```python
formatted_times_pyrometer = []
for a_time in pyrometer_times:
    current_time_second = float((a_time.split(':'))[-1])
    current_time_minute = float((a_time.split(':'))[-2])
    #print(current_time_minute)
    #print(current_time_second)
    current_time = (current_time_minute * 60) + current_time_second
    formatted_times_pyrometer.append(current_time)
```

```python
time_period_raw = 0.001
time_preiod_pyrometer = 0.0294

time_ratio_pyro_to_raw = time_preiod_pyrometer/time_period_raw
time_ratio_pyro_to_raw_int = int(time_ratio_pyro_to_raw)
#print(time_ratio_pyro_to_raw_int)

# If this value gets bigger than 1, we can skip 1 extra frame, and set it again to zero
Extra_frame = 0.0

# Create an empty list to hold the synchronized data
raw_voltage_sync = []
raw_voltage_sync_frames = []

current_index = 0
for current_frame in range(np.shape(calibrated_TProc_sync_zero)[0]):
    
    raw_voltage_sync.append(raw_voltage_sync_zero[current_index])
    raw_voltage_sync_frames.append(current_index)
    
    current_index += time_ratio_pyro_to_raw_int
    Extra_frame += (time_ratio_pyro_to_raw - time_ratio_pyro_to_raw_int)
    if Extra_frame>=1:
        current_index += 1
    if current_frame == (np.shape(calibrated_TProc_sync_zero)[0])-1:
        print(f'The last UNPROCESSED frame number in the raw voltage data is: {current_index}')
```
